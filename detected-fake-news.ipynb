```python
from sklearn.model_selection import train_test_split
from transformers import BertTokenizer
import torch

df['Processed_Text'] = df['Processed_Text'].fillna('')
# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(df['Processed_Text'], df['label'], test_size=0.2, shuffle = True, random_state=42)

# Инициализация токенизатора BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

# Токенизация данных
def encode_data(texts, tokenizer, max_length):
    input_ids = []
    attention_masks = []

    for text in texts:
        encoded_dict = tokenizer.encode_plus(
                            text,
                            add_special_tokens = True,
                            max_length = max_length,
                            padding = 'max_length',
                            truncation=True,
                            return_attention_mask = True,
                            return_tensors = 'pt'
                       )

        input_ids.append(encoded_dict['input_ids'])
        attention_masks.append(encoded_dict['attention_mask'])

    return torch.cat(input_ids, dim=0), torch.cat(attention_masks, dim=0)

# Установка максимальной длины последовательностей
max_length = 256

# Кодирование тренировочных данных
train_inputs, train_masks = encode_data(X_train, tokenizer, max_length)
test_inputs, test_masks = encode_data(X_test, tokenizer, max_length)

# Преобразование меток в тензоры
train_labels = torch.tensor(y_train.values)
test_labels = torch.tensor(y_test.values)

# Создание DataLoader
from torch.utils.data import DataLoader, TensorDataset, RandomSampler, SequentialSampler

batch_size = 16

# Создание TensorDataset
train_data = TensorDataset(train_inputs, train_masks, train_labels)
test_data = TensorDataset(test_inputs, test_masks, test_labels)

# Создание DataLoader для тренировочных данных
train_dataloader = DataLoader(
            train_data,
            sampler = RandomSampler(train_data),
            batch_size = batch_size
        )

# Создание DataLoader для тестовых данных
test_dataloader = DataLoader(
            test_data,
            sampler = SequentialSampler(test_data),
            batch_size = batch_size
        )
```

```python
from transformers import BertForSequenceClassification, AdamW, get_linear_schedule_with_warmup
import torch
import numpy as np
from tqdm import tqdm  # Импортируем tqdm

# Инициализация модели BERT для классификации
model = BertForSequenceClassification.from_pretrained(
    "bert-base-multilingual-cased",
    num_labels=2,  # Установите количество классов
    output_attentions=False,
    output_hidden_states=False,
)

# Использование GPU, если доступно
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Оптимизатор и планировщик обучения
optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8)
epochs = 4
total_steps = len(train_dataloader) * epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

# Функция для вычисления точности
def flat_accuracy(preds, labels):
    pred_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()
    return np.sum(pred_flat == labels_flat) / len(labels_flat)

# Обучение модели
for epoch_i in range(0, epochs):
    print(f'Эпоха {epoch_i + 1}/{epochs}')

    model.train()
    total_loss = 0

    # Используем tqdm для отображения прогресса обучения
    for step, batch in enumerate(tqdm(train_dataloader, desc="Обучение")):
        b_input_ids = batch[0].to(device)
        b_input_mask = batch[1].to(device)
        b_labels = batch[2].to(device)

        model.zero_grad()

        outputs = model(b_input_ids, token_type_ids=None, attention_mask=b_input_mask, labels=b_labels)
        loss = outputs.loss
        logits = outputs.logits

        total_loss += loss.item()

        loss.backward()
        optimizer.step()
        scheduler.step()

    avg_train_loss = total_loss / len(train_dataloader)
    print(f"Средняя потеря обучения: {avg_train_loss}")

# Тестирование модели
print("Тестирование модели...")

model.eval()
eval_accuracy = 0
nb_eval_steps = 0

# Используем tqdm для отображения прогресса тестирования
for batch in tqdm(test_dataloader, desc="Тестирование"):
    b_input_ids = batch[0].to(device)
    b_input_mask = batch[1].to(device)
    b_labels = batch[2].to(device)

    with torch.no_grad():
        outputs = model(b_input_ids, token_type_ids=None, attention_mask=b_input_mask)

    logits = outputs.logits
    logits = logits.detach().cpu().numpy()
    label_ids = b_labels.to('cpu').numpy()

    eval_accuracy += flat_accuracy(logits, label_ids)
    nb_eval_steps += 1

print(f"Точность тестирования: {eval_accuracy / nb_eval_steps}")
```
